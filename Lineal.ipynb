```python
import matplotlib.pyplot as plt
from sklearn import linear_model
import seaborn as sns
import pandas as pd
import numpy as np
import os
```

```python
lineal = pd.read_csv("../anderson/data.csv", sep=",")
```

```python
lineal
```

```python
lineal.plot.scatter(x="metro", y="precio")
plt.show()
```

```python
regresion = linear_model.LinearRegression()
```

```python
metros = lineal["metro"].values.reshape((-1,1))
```

```python
valor = regresion.fit(metros, lineal["precio"])
```

```python
print("Interseccion (b)", valor.intercept_)
```

```python
print("Pendiente (m)", valor.coef_)
```

```python
entrada= [[5.0], [12.5], [22.5], [55.0]]
predicciones = valor.predict(entrada)
print(predicciones)
```

```python
lineal.plot.scatter(x="metro", y="precio", label='Valores Iniciales')
plt.scatter(entrada, predicciones, color='red')
plt.plot(entrada, predicciones, color='gold', label="Linea de regresin") 
plt.xlabel('metro') 
plt.ylabel('precio')
plt.legend() 
plt.show()
```
